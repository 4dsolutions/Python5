## Constructing a Binomial Distribution

What other functions do we remember from high school?  

Likely lines, polynomials, exponential and logarithmic curves.

Lets build up a dataset using permutations of Left (L) and (R), 10 of them in a sequence.  Imagine flipping a fair coin to get essentially the same examples.  "LLRRLRRLLR" would be a member of our set.  We won't be constructing every possible sequence.  Think of "GATACA" as a string of DNA.

The plan is to count the number of lefts (Ls) in 5000 samples and then bin them in buckets of 0 through 10, for display in a histogram.  What I expect is roughly a binomial distribution, with all Ls, and no Ls, each equally unlikely, and therefore at the two tails.  The "most average" sample will have about the same of each.

````python
import random
from random import choice
````

````python
random.seed = 42  # start point in the orbit
````

````python
def galton_board(the_choices = ['L','R'], the_len = 10):
    random_string = ""
    for _ in range(the_len):
        random_string += choice(the_choices)
    return random_string
````

````python
sample = galton_board()
sample
````

````python
L_counts = [galton_board().count("L") for _ in range(5000)]
````

Binning may be a little tricky, as it's the upper and lower edges you need to specify.  You may want those at half-way points between whole numbers.  That's how I design things in the code cell below.

````python
bins = np.arange(12) - 0.5 # leftmost bin: between -0.5 and 0.5, then .5 and 1.5 etc.
plt.xticks(range(11))
plt.xlim([-1, 11])
plt.hist(L_counts, bins=bins)
plt.title("Frequency of Ls")
plt.xlabel("Binned Samples")
_ = plt.ylabel("L counts")
````

Here's what the Gaussian looks like over the domain we used for out trig functions up to i.e. ```[-10, 10]```. 

This domain has little to do with the above histogram, which bins the counts for the letter L.  

Note that the standard normal curve, as it's called, is only allowed to:

    * have a total area of 1 between itself and the X axis
    * have a standard deviation of 1, meaning over 68% of the samples occur within 1 of the mean
    * have a mean at 0
    
To make a Gaussian compatible with an actual data set, it helps to already know the mean and standard deviation, in which case ```st.norm.pdf``` is standing by to take those values, as ```loc``` and ```scale``` respectively.

For reference, 68.27% of values in a normal distribution lie within 1 SD (+/-) of the mean, 95.45% lie within 2 SD's, and 99.73% lie within 3 SD's.
